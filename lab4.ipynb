```python
# Parent class for mental health professional
class MentalHealthProfessional:
    def __init__(self, name, specialization):
        self.name = name
        self.specialization = specialization

    def introduce(self):
        print(f"I am {self.name}, a {self.specialization}.")

# Child class inheriting from MentalHealthProfessional
class Psychiatrist(MentalHealthProfessional):
    def __init__(self, name, specialization, experience):
        super().__init__(name, specialization)
        self.experience = experience

    def diagnose(self, patient):
        print(f"{self.name} is diagnosing {patient}.")

# Child class inheriting from MentalHealthProfessional
class Psychologist(MentalHealthProfessional):
    def __init__(self, name, specialization, therapy_type):
        super().__init__(name, specialization)
        self.therapy_type = therapy_type

    def provide_therapy(self, patient):
        print(f"{self.name} is providing {self.therapy_type} therapy to {patient}.")

class conditions():
    def __init__(self,condition_name,type_name,therapy_type):
        self.condition_name = condition_name
        super().__init__(type_name,therapy_type)
    def condition_type(self):
        print(f"{self.condtion_name} is need to be treated")

    

    
class type():
    def __init__(self, type_name,therapy_type):
        self.type_name = type_name
        self.therapy_type = therapy_type
        

    def provide_therapy(self, patient):
        print(f"{self.type_name} need {self.therapy_type} therapy to {patient}.")


class Psychotherapist(conditions, type):
        
    def __init__(self,name, type_name,condition_name, specialization, experience, therapy_type):
        super().__init__(condition_name,type_name,therapy_type)
        self.name = name
        self.specialization = specialization
        self.experience = experience
        

    def introduce(self):
        print(f"I am {self.name}, a Psychotherapist with {self.experience} years of experience "
              f"and specialized in {self.specialization}. I provide {self.therapy_type} therapy.")

    def prescribe_medication(self, patient):
        print(f"{self.name} is prescribing medication to {patient}.")   
       

  

# Creating objects and demonstrating the inheritances
professional = MentalHealthProfessional("Dr. Borah", "Mental Health Professional")
professional.introduce()

psychiatrist = Psychiatrist("Dr. Johnson", "Psychiatrist", 10)
psychiatrist.introduce()
psychiatrist.diagnose("John")

psychologist = Psychologist("Dr. Davis", "Psychologist", "Cognitive Behavioral Therapy")
psychologist.introduce()
psychologist.provide_therapy("Jane")

psychotherapist = Psychotherapist("Dr. Anderson", "Psychotherapist", 5, "Psychoanalytic Therapy","dfcvghjfg","fghjdfghf")
psychotherapist.introduce()
psychotherapist.provide_therapy("Emily")
psychotherapist.prescribe_medication("Sarah")
```
